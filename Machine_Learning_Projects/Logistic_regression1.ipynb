```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
insurence=pd.read_csv("D:/Data MLE/insurance_data.csv")
insurence.head()

```

```python
sns.set_style('whitegrid')
sns.scatterplot(data=insurence,x='age',y='bought_insurance')
plt.plot(insurence.age,insurence.bought_insurance)
```

```python
insurence.info()
```

```python
insurence['bought_insurance'].value_counts()
```

```python
x=insurence[['age']]
y=insurence.bought_insurance
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```

```python
y_test
```

```python
model.score(x_test,y_test)
```
