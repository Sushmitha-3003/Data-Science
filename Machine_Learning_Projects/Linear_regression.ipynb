```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
```

```python
df=pd.read_csv("D:/Data MLE/area_price.csv")
df
```

```python
plt.scatter(df.area,df.price,color='red',marker='*')
plt.xlabel('area(sqrt)')
plt.ylabel('price(US$)')
plt.show()
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[4000]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg_predict=reg.coef_*4000+reg.intercept_
reg_predict
```

```python
plt.scatter(df.area,df.price,color='red',marker='*')
plt.xlabel('area(sqrt)')
plt.ylabel('price(US$)')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
plt.show()
```
